```python
stop_word = stopwords.words('english')
stop_word
```

```python
def clean(text):
    text = re.sub(r'http\S+', " ", text)
    text = re.sub(r'@\w+',' ',text)
    text = re.sub(r'#\w+', ' ', text)
    text = re.sub(r'\d+', ' ', text)
    text = re.sub('r<.*?>',' ', text)
    text = text.split()
    text = " ".join([word for word in text if not word in stop_word])
    
      
    return text
```

```python
test=df1_test['comment'].str.lower()
test
x_test=test
y_test=df1_test['rating']
print(x_test)
print(y_test)
```

```python
model_nb = Pipeline([
    ('count_vectorizer', CountVectorizer(lowercase = True, stop_words = stopwords.words('english'))), 
    ('tfidf_transformer',  TfidfTransformer()), #weighs terms by importance to help with feature selection
    ('classifier', MultinomialNB(alpha=3)) ])
model_nb.fit(x_train,y_train.astype('int'))
y_pred=model_nb.predict(x_test)
accuracy=accuracy_score(y_test.astype('int'),y_pred, normalize=True) * float(100)
print('\n****Test accuracy is',(accuracy))
```
